```python
#zad 14
def kalk(w1,dz,w2):
    try:
        if dz=='+':
            return w1+w2
        elif dz=='-':
            return w1-w2
        elif dz=='*':
            return w1*w2
        elif dz=='/':
            return w1/w2
        else: return print("złe dzialanie")
    except:
        print("błąd")
kalk('w','-',3)
```

```python
#zad 15
def kalk_zab(w1,dz,w2):
    try:
        if dz=='+':
            return w1+w2
        elif dz=='-':
            return w1-w2
        elif dz=='*':
            return w1*w2
        elif dz=='/':
            if w2==0:
                return print("dzielnie przez 0!")
            else: return w1/w2
        else: return print("złe dzialanie")
    except:
        print("błąd")
kalk_zab('a','/',2)
```
